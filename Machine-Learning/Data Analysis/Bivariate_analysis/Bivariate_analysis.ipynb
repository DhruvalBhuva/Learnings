## 3. Box Plot (Numerical - Categorical)

```python
# sns.boxplot(x=titanic['Sex'])
# sns.boxplot(x=titanic['Sex'],y=titanic['Age'])
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

## 4. Distplot (Numerical - Categorical)

```python
# Distplot don't have the hue parameter, so we need to make custom dataframe
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

## 5. HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
print(pd.crosstab(titanic['Pclass'],titanic['Survived']))
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
# It giives percentage of survived people
(titanic.groupby('Pclass').mean()['Survived']*100)

# (titanic.groupby('Embarked').mean()['Survived']*100)
# (titanic.groupby('Sex').mean()['Survived']*100)

# (titanic.groupby('Embarked').mean()['Survived']*100).plot(kind="bar")
```
